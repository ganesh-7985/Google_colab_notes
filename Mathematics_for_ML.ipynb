#Vector Operations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set() # sets the basics requirement to represent a vector/graph
```

Plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy' ,angles = 'xy', scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy' ,angles = 'xy', scale=1,color='b')
plt.quiver(0,0,-3,5, scale_units='xy' ,angles = 'xy', scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Vector Addition

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum = vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy' ,angles = 'xy', scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy' ,angles = 'xy', scale=1,color='y')
plt.quiver(0,0,5,1, scale_units='xy' ,angles = 'xy', scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Vector Subtraction

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
diff = vector_1-vector_2
print(diff)
```

```python
plt.quiver(0,0,2,3, scale_units='xy' ,angles = 'xy', scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy' ,angles = 'xy', scale=1,color='y')
plt.quiver(0,0,-1,5, scale_units='xy' ,angles = 'xy', scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot product of two vectors

```python
a=np.array([2,3])
b=np.array([4,4])
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

Cross product of two vectors

```python
a=np.array([2,3])
b=np.array([4,4])
a_cross_b = np.cross(a,b)
print(a_cross_b)
```

```python

# create two matrices with random values
matrix_1 = np.random.randint(10, size=(3,3))
matrix_2 = np.random.randint(20, size=(3,3))
print(matrix_1)
print('')
print(matrix_2)
```

```python
sum = np.add(matrix_1, matrix_2)
print(sum)
```

```python
diff = np.subtract(matrix_1, matrix_2)
print(diff)
```

```python
product = np.dot(matrix_1, matrix_2)
print(product)
```

```python
cross_product = np.multiply(matrix_1, matrix_2)
print(cross_product)
```
