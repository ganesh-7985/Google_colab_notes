```python
# LIST
# List also know as array in cpp list can contain duplicates.list are represented in the square brackets
list_1 = [1,2,4,3,4,5]
list_1.append(6)

list_2=[4,4,5,3,2]
list_3=list_1+list_2
list_3.append("hello")
print(list_3)

del list_3[12]
print(list_3)
```

List Vs Numpy -Time Taken

```python
from time import process_time
```

```python
python_list = [i for i in range(10000)]
start_time = process_time()
python_list = [i+5 for i in python_list]
end_time = process_time()
print(end_time-start_time);
```

```python
# using Numpy library
np_array = np.array([i for i in range(10000)])
start_time = process_time()
np_array +=5
end_time = process_time()
print(end_time-start_time);
```

```python
c = np.random.randint(10,100,(5,5))
print(c)
print(c.size)   # No. of elements
print(c.shape)  # array dimension
print(c.ndim)   # no of dimensions
print(c.dtype)  # datatypes
type(c)
```

Handling the missing values in the Dataset

* Imputation:method in which we replace null value with mean,median or mode values

examples:

dataset['salary'].fillna(dataset['salary'].mean(),inplace=True) or
dataset['salary'].fillna(dataset['salary'].median(),inplace=True) or
dataset['salary'].fillna(dataset['salary'].mode(),inplace=True)

* Dropping:where we drop the row of that value where it is null

examples:

dataset.dropna(how='any')

**Data Standardization**

the process of converting data to a common format or common range to enable users to process and analyze it

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
# loading the data to pandas to dataframe
df = pd.DataFrame(dataset.data,columns = dataset.feature_names)
print(df)
```

```python
df.head()
```

```python
X = df
Y = dataset.target
```

splitting the data into training data and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# standarized the data
# if standard deviation = 1 then it already standardized if we need we to standardize it
print(dataset.data.std())
# it value is around 228 it is not standardized
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_standardized = scaler.transform(X_train)
```

```python
print(X_train_standardized)
```

```python
print(X_train_standardized.std())
```
