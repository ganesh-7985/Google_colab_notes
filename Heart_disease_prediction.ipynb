```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and PreProcessing

```python
# Load the csv data to the pandas
data = pd.read_csv('/content/heart_data.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
#checking for the missing values
data.isnull().sum()
```

```python
# their no missing values so we can all 303 rows to train and test the model
#statistical measures of the data
data.describe()
```

```python
# checking the distribution of the Target variable
data['target'].value_counts()
```

1 --> Defective Hearts

0 --> Normal Hearts

```python
# splitting the features(input values) and target(output values)
X=data.drop(columns='target', axis=1)
Y=data['target']
print(X)
print('')
print(Y)
```

```python
# splitting the data into training data and testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
#LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print(training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print(testing_data_accuracy)
```

predict accoding to input

```python
input = (41,0,1,130,204,0,0,172,0,1.4,2,0,2)
input_data_numpy = np.asarray(input)
input_data_reshaped = input_data_numpy.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)
```
